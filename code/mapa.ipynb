```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile of Guatemala City and zones
guate = gpd.read_file('/Users/rudikrompich/Git/analisis_tse_elecciones_2023/map_graph/municipios_GTM.shx')

guate.head(5)
```

```python

# Plot the map
fig, ax = plt.subplots(figsize=(10, 10))
guate.plot(ax=ax)

# Add title and labels
ax.set_title('Map of Guatemala City and Zones')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```
